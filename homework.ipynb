```python
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np

def create_pixel_mask(img_path, h, x_min, x_max, y_min, y_max):
    img = imread(img_path)

    width = (x_max - x_min) // h
    height = (y_max - y_min) // h
    
    img_array = np.array(img)
    
    for i in range(h):
        for j in range(h):
            x_start = x_min + i * width
            y_start = y_min + j * height
        
            region = img_array[y_start : y_start + height, x_start : x_start + width]
        
            region_color = (int(np.mean(region[:, :, 0])),  int(np.mean(region[:, :, 1])),  int(np.mean(region[:, :, 2])))

            img_array[y_start : y_start + height, x_start : x_start + width] = np.full_like(region, region_color)

    plt.imshow(img_array)
    plt.show()

img_path = 'https://i.pinimg.com/564x/19/6d/cb/196dcb6b701056eeacb2aff778562cf5.jpg'
x_max, x_min = 400, 100
y_max, y_min = 300, 150          
h = 15

orig_img = imread(img_path)
plt.imshow(orig_img)
plt.show()

create_pixel_mask(img_path, h, x_min, x_max, y_min, y_max)
```
